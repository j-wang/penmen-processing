Preprocessing
----------------

```python
pen = scipy.misc.imread('pentest.jpg')

def map_on_rgb(arr, f):
    a = np.reshape(arr, (-1, 3))
    for i, v in enumerate(a):
        a[i] = f(v)
    return arr

def black_and_white(rgb_arr, threshold=105):
    r, g, b = rgb_arr
    if r < threshold:
        return [0, 0, 0]
    else:
        return [255, 255, 255]

map_on_rgb(pen, black_and_white)

pylab.imshow(pen)  # obviously, the image representation is mutable...
pylab.show()
scipy.misc.imsave('test/pentester.png', pen)
```

Finding Groups of Penmen
-------------------------

```python
def flood_find(arr):
    def flood(coord, identif):
        x, y = coord
        rgb = arr[x][y]
        if rgb[0] != 0:
            return
        else:
            rgb[0] = identif
            neigh = [(i, j) for i in [x, x - 1, x + 1] for j in [y, y - 1, y + 1]]
            neigh.remove((x, y))
            for cell in neigh:
                try:
                    flood(cell, identif)
                except IndexError:
                    pass
    
    row, col, _ = arr.shape
    for r in range(row):
        for c in range(col):
            flood((r, c), hash((r, c)))
            
test_arr = np.array([[[255, 255, 255], [255, 255, 255], [0, 0, 0]],
                     [[0, 0, 0], [255, 255, 255], [0, 0, 0]],
                     [[0, 0, 0], [255, 255, 255], [0, 0, 0]]])

flood_find(test_arr)
print test_arr
```

```python
def colorize(arr):
    row, col, _ = arr.shape
    for r in range(row):
        for c in range(col):
            rgb_cell = arr[r][c]
            red, gre, blu = rgb_cell
            if gre != 255:
                rgb_cell[0] = hash(red) % 255
                rgb_cell[1] = hash((red, red)) % 255
                rgb_cell[2] = hash((red, red, red)) % 255
```

```python
pen = scipy.misc.imread('pentest.jpg')
bw = lambda x: black_and_white(x, threshold=105)
map_on_rgb(pen, bw)
flood_find(pen)
colorize(pen)
scipy.misc.imsave('test/pentester_grouped.png', pen)
pylab.imshow(pen)
pylab.show()
```

```python
# pen = scipy.misc.imread('pentest.jpg')
# bw = lambda x: black_and_white(x, threshold=105)
# map_on_rgb(pen, bw)

# min, if negative add constant to read 0 to all
# if min is positive, subtract to all to get to 0

def grab_groups(arr):
    penman = []
    groups = []
    row, col, _ = arr.shape
    append = groups.append  # for performance
    
    def grab(coord):
        x, y = coord
        if x < 0 or y < 0:
            raise IndexError
        rgb_cell = arr[x][y]
        if rgb_cell[2] != 0:
            return
        else:
            penman.append((x, y))
            rgb_cell[2] = 1  # mark visited
            neigh = [(i, j) for i in [x, x - 1, x + 1] for j in [y, y - 1, y + 1]]
            neigh.remove((x, y))
            for cell in neigh:
                try: 
                    grab(cell)
                except IndexError:
                    pass

    def normalize(grabbed):
        min_x = min(grabbed, key=lambda x: x[0])[0]
        min_y = min(grabbed, key=lambda x: x[1])[1]
        return [frozenset([(x[0] - min_x, x[1] - min_y) for x in grabbed]), grabbed]
    
    for r in range(row):
        for c in range(col):
            grab((r, c))
            if penman != []:
                append(normalize(penman))
                penman = []
                
    return groups

test_arr = np.array([[[255, 255, 255], [255, 255, 255], [0, 0, 0]],
                     [[0, 0, 0], [255, 255, 255], [0, 0, 0]],
                     [[0, 0, 0], [255, 255, 255], [0, 0, 0]]])

print grab_groups(test_arr)
```

Matching Penmen
-----------------

```python
# don't compare if pixels aren't similar

def find_similar(grabbed_groups, threshold=20, ignore=[set()]):
    def fuzzy(s):
        return {(x + i, y + j) for x, y in s 
                               for i, j in [(-1, -1), (-1, 0), (-1, 1), 
                                            (0, -1), (0, 0), (0, 1), 
                                            (1, -1), (1, 0), (1, 1)]}
    groups = []
    ignored = {}
    combos = itertools.combinations(grabbed_groups, 2)
    for penman1, penman2 in combos:
        set1, set2 = penman1[0], penman2[0]
        if set1 not in ignored and set2 not in ignored:
            ignore_penman = False
            if ignore != [set()]:
                for item in ignore:
                    fuzzed = fuzzy(item)
                    set1_diff = 100 * len(set1 - fuzzed) / len(set1 | item)
                    set2_diff = 100 * len(set2 - fuzzed) / len(set2 | item)
                    if set1_diff < threshold:
                        ignore_penman = True
                        ignored.update(set1)
                    elif set2_diff < threshold:
                        ignore_penman = True
                        ignored.update(set2)
            if not ignore_penman:
                difference1 = set2 - fuzzy(set1)
                difference2 = set1 - fuzzy(set2)
                diffs = 100 * len(difference1 | difference2) / len(set1 | set2)
                if diffs < threshold:
                    groups.append((diffs, (penman1[1], penman2[1])))
    return sorted(groups)
```

```python
pen = scipy.misc.imread('pentest.jpg')
map_on_rgb(pen, black_and_white)
similarity_list = find_similar(filter(lambda x: len(x[0]) > 50, grab_groups(pen)))
print similarity_list
```

```python
def highlight_top(img, coord_list):
    length = len(coord_list)
    for i, item in enumerate(coord_list):
        blob1, blob2 = item[1]
        for coord in blob1 + blob2:
            x, y = coord
            img[x][y] = [255, 255 * i/(length - 1), 0]
            
pen = scipy.misc.imread('pentest.jpg')
highlight_top(pen, similarity_list[:50])
pylab.imshow(pen)
pylab.show()
```

Importance of Fuzziness
------------------------

```python
pentest_easy = scipy.misc.imread('deliberate_copy_simple.jpg')
bw = lambda x: black_and_white(x, threshold=105)
map_on_rgb(pentest_easy, bw)

grab_groups_easy = grab_groups(pentest_easy)
sim_list_easy = find_similar(filter(lambda x: len(x[0]) > 60, grab_groups_easy))
highlight_top(pentest_easy, sim_list_easy)
scipy.misc.imsave('test/pentest_easy_grouped.png', pentest_easy)
pylab.imshow(pentest_easy)
pylab.show()

filtered_group = filter(lambda x: len(x[0]) > 60, grab_groups_easy)
print len(sim_list_easy)
print len(filtered_group)
print sorted(filtered_group[0][0])
print '\n\n'
print sorted(filtered_group[2][0])
print '\n\n'

print 'This is pretty bad for pixel perfect copies. There was some sort of noise that added a pixel.'
print sorted(filtered_group[0][0] ^ filtered_group[2][0])
print '\n\n'

print 'Fuzziness deals with this problem.'
def futz(s): return {(x-i,y-j) for x,y in s for i in [-1,0,1] for j in [-1,0,1]}
print 100 * len(filtered_group[0][0] - futz(filtered_group[2][0])) / len(filtered_group[0][0] | filtered_group[2][0])
print sorted(filtered_group[0][0] - futz(filtered_group[2][0]))
```

```python
pen1 = scipy.misc.imread('penmen1copy.png')
heads = scipy.misc.imread('heads.jpg')
bw = lambda x: black_and_white(x, threshold=105)
map_on_rgb(heads, bw)
map_on_rgb(pentest, bw)

headset = [pair[0] for pair in grab_groups(heads)]

map_on_rgb(pen1, bw)
sim_list = find_similar(filter(lambda x: len(x[0]) > 60, grab_groups(pen1)), ignore=headset)
highlight_top(pen1, sim_list[:len(sim_list) / 2])
scipy.misc.imsave('test/pen1_filter_find.png', pen1)
pylab.imshow(pen1)
pylab.show()
```

```python
def output_similar(img, coord_list):
    length = len(coord_list)
    for i, item in enumerate(coord_list):
        blob1, blob2 = item[1]
        c1 = hash(blob1[0]) % 255
        c2 = hash(blob2[0]) % 255
        c3 = hash((blob1[0], blob2[0])) % 255
        for coord in blob1 + blob2:
            x, y = coord
            img[x][y] = [c1, c2, c3]
            
def kill_noncolored(rgb_arr):
    r, g, b = rgb_arr
    if r + g + b == 0:
        return [255, 255, 255]
    else:
        return rgb_arr
            
pen_results = copy.deepcopy(pen1)
output_similar(pen_results, sim_list[:len(sim_list) / 2])
map_on_rgb(pen_results, kill_noncolored)
scipy.misc.imsave('test/pen1_filter_find_grouped.png', pen_results)

pylab.imshow(pen_results)
pylab.show()
```
